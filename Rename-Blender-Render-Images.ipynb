# General notes
All images rendered from Blender should be renamed for easier importation later. By default, renders from blender are named after the frame number (e.g. '420.png'). Images should be ranamed into format 'piece-index_chess-field_rotation.png'. Image name of white king at chess filed number 44 and rotation 8/16 is then '3_44_8.png'.

```python
import os
import glob
```

Define the name of the folder where the Blender renders are located, the extension of the images ('.png') and the appropriate piece name. This code renames images for one piece at the time. For this reason, in the Blender renders folder, there should only be images for one piece. The number of different angles at which the piece was rendered at each chess board field should be defined aswell. When selecting piece_name, use this table: 

| INDEX | PIECE NAME | CLASS NAME |
| :-: | :-: | :-: |
| 0 | bishop black | b |
| 1 | bishop white | B |
| 2 | king black | k |
| 3 | king white | K |
| 4 | knight black | n |
| 5 | knight white | N |
| 6 | pawn black | p |
| 7 | pawn white | P |
| 8 | queen black | q |
| 9 | queen white | Q |
| 10 | rook black | r |
| 11 | rook white | R |

```python
pathrender='blender_render' # Blender renders folder
image_extension='.png' # .png is default, since the images include the Alpha channel

piece_index='0' # Use table help

field_angles = 16 # Number of different angles
```

A list of names is created (first loop), then the images are renamed (second loop).

```python
all_names=[]

for i in range(1,65):
    for j in range(1,field_angles+1):
        name=piece_index+'_'+str(i)+'_'+str(j)+image_extension
        all_names.append(name)

index=0
for filename in glob.glob(os.path.join(pathrender,'*')):
    os.rename(filename,os.path.join(pathrender,all_names[index]))
    print(filename,'>>>WAS RENAMED TO>>>',os.path.join(pathrender,all_names[index]))
    index+=1
```
